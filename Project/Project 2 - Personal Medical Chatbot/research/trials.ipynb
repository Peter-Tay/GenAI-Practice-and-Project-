```python
from langchain.document_loaders import PyPDFLoader, DirectoryLoader
from langchain.text_splitter import RecursiveCharacterTextSplitter
```

```python
#Extract text frrrom pdf files
def load_pdf_files(data_path):
    loader = DirectoryLoader(
        data_path,
        glob="*.pdf",
        loader_cls=PyPDFLoader
    )

    documents = loader.load()
    return documents
```

```python
extracted_data = load_pdf_files("data")
```

```python
extracted_data
```

```python
len(extracted_data)
```

```python
from typing import List
from langchain.schema import Document

def filter_to_simplify_docs(docs: List[Document]) -> List[Document]:
    """
    Only keep metadata 'source' and original page_content
    """
    simp_docs: List[Documeent] = []
    for doc in docs:
        src = doc.metadata.get("source")
        simp_docs.append(
            Document(
                page_content=doc.page_content,
                metadata={"source": src},
            )
        )
    return simp_docs
```

```python
simp_docs = filter_to_simplify_docs(extracted_data)
```

```python
simp_docs
```

```python
#split documents into smaller chunks
def text_split(docs):
    text_splitter = RecursiveCharacterTextSplitter(
        chunk_size=500,
        chunk_overlap=20,
        length_function=len
    )
    texts_chunk = text_splitter.split_documents(docs)
    return texts_chunk
```

```python
texts_chunk = text_split(simp_docs)
print(f"Number of chunks: {len(texts_chunk)}")
```

```python
texts_chunk
```

```python
from langchain.embeddings import HuggingFaceEmbeddings

def get_embeddings_model():
    """
    Return HuggingFace embedddings model.
    """
    model_name = "sentence-transformers/all-MiniLM-L6-v2"
    embeddings = HuggingFaceEmbeddings(
        model_name=model_name,
    )
    return embeddings
```

```python
embedding = get_embeddings_model()
```

```python
from langchain_openai import ChatOpenAI

chat_model = ChatOpenAI(model="gpt-4o")
```

```python
from langchain.chains import create_retrieval_chain
from langchain.chains.combine_documents import create_stuff_documents_chain
from langchain_core.prompts import ChatPromptTemplate
```

```python
system_prompt = (
    "You are an assistant for question-answering tasks. "
    "Use the following pieces of retrieved context to answer "
    "the question. If you don't know the answer, say that you"
    "don't know. Use three sentences maximum and keeep the"
    "answer concise."
    "\n\n"
    "{context}"
)

prompt = ChatPromptTemplate.from_messages(
    [
        ("system", system_prompt),
        ("human", "{input}")
    ]
)
```

```python
question_answer_chain = create_stuff_documents_chain(chat_model, prompt)
rag_chain = create_retrieval_chain(retriever, question_answer_chain)
```

```python
response = rag_chain.invoke({"input": "what is Acromegaly and gigantism?"})
print(response["answer"])
```
